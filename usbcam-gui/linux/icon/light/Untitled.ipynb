```python
import cv2
import matplotlib.pyplot as plt


f = "close.png"

im = cv2.imread(f, -1)
#im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
fig, ax = plt.subplots()
ax.imshow(im)
print(im.shape)
```

```python
b, g, r = cv2.split(im)
```

```python
new_b = 255 - b
new_g = 255 - g
new_r = 255 - r
print(new_b)
print(new_g)
print(new_r)

after = cv2.merge((new_b, new_g, new_r))
print(after)
cv2.imwrite("after.png", after)
```

```python
f2 = "after.png"

im = cv2.imread(f2, -1)
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
fig, ax = plt.subplots()
ax.imshow(im)
print(im.shape)
```
